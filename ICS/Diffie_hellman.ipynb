```python
import pyDH
d1 = pyDH.DiffieHellman()
d2 = pyDH.DiffieHellman()
d1_priv_key = d1.get_private_key()
d2_priv_key = d2.get_private_key()
d1_pubkey = d1.gen_public_key()
d2_pubkey = d2.gen_public_key()
d1_sharedkey = d1.gen_shared_key(d2_pubkey)
d2_sharedkey = d2.gen_shared_key(d1_pubkey)
d1_sharedkey == d2_sharedkey


print("D1")
print("Private key:", d1_priv_key)
print("Public key:", d1_pubkey)
print("Shared key:", d1_sharedkey)

print("D2")
print("Private key:", d2_priv_key)
print("Public key:", d2_pubkey)
print("Shared key:", d2_sharedkey)
```
